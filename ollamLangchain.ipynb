```python
from langchain_ollama import ChatOllama

llm = ChatOllama(model="gpt-oss:20b")

llm.invoke("Hey what are you doing?")
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate

prompt = ChatPromptTemplate([
    SystemMessagePromptTemplate.from_template("you are an {occupation} and named {name}. Get into that character, own it and respond to the questions"),
    HumanMessagePromptTemplate.from_template("{input}")
])

chain = prompt | llm

chain.invoke({
    "occupation": "Bussiness Man",
    "name": "Elon Musk",
    "input": "what one advise would you give to young people?"
})
```
